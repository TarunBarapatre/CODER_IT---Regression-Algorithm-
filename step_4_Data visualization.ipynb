```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
year = [2014,2015,2016,2017,2018]
sales = [1234,4344,4454,3222,3343]
profit = [123,443,445,432,222]

```

```python
plt.plot(year, sales, 'green')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales vs Year')
```

```python
customer = ['Customer1','Customer2','Customer3','Customer4','Customer5']
```

```python
plt.scatter(sales,profit, color='red', marker='*')
```

```python
plt.plot(year,sales,'red')
plt.plot(year, profit,'blue')
```

```python
plt.axes([1,1,0.9,0.9])
plt.plot(year, sales,'red')
plt.title('Year and Sales')
plt.axes([0.05,1.0,0.9,0.9])
plt.plot(year, profit,'blue')
plt.title('Year and Profit')
```

```python
plt.subplot(2,1,1)
plt.plot(year, sales,'red')
plt.title('Year and Sales')
#plt.xlim((2014,2015))
#plt.ylim((1500,2500))
#plt.axis((2014,2015,1500,2500))
plt.subplot(2,1,2)
plt.plot(year, profit,'blue')
plt.title('Year and Profit')
plt.tight_layout()
```

```python
plt.scatter(sales,profit)
plt.annotate('2014',xy=(1234,123))
# Give the name of plot indiviualy
plt.style.use('classic')
```
